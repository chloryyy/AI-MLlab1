```python
import pandas as pd
pd.set_option('future.no_silent_downcasting', True)
df = pd.read_csv("train.csv")
df.info()
```

```python
num_columns = df.select_dtypes(include=['number']).columns
df[num_columns] = df[num_columns].fillna(df[num_columns].median())

categ_columns = df.select_dtypes(include=['object', 'bool']).columns
df[categ_columns] = df[categ_columns].fillna(df[categ_columns].mode().iloc[0])
df.info()
```

```python
missing_values = df.isnull().sum()
missing_values
```

```python
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
scaler1 = MinMaxScaler()
scaler2 = StandardScaler()
df["Age"] = scaler1.fit_transform(df[["Age"]])
df["FoodCourt"] = scaler2.fit_transform(df[["FoodCourt"]])
```

```python
df = pd.get_dummies(df, columns=["HomePlanet"], drop_first = True)
```

```python
df.head()
```

```python
df.to_csv("traincompleted.csv", index = False)
```
